```python
import pandas as pd

#df = dataframe
#error_bad_lines = ignorar linhas com erro, sep = separador, no caso, separador da planilha q nesse arquivo aqui é ;, por padrão é ,
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dados/folhapagamento_202304.csv", error_bad_lines=False, sep=";")

#Visualizando as 5 primeiras linhas
df.head()

#mudando nome das colunas
df = df.rename(columns={"CODORGAO": "Codigo do Orgao", "NOMESERVIDOR": "Nome do servidor"})

#só colocar o numero de linhas que você quer dentro dos parenteses. Se não passar nenhum numero, ele mostra as 5 primeiras
df.head(10)
```

```python
#retorna os nomes unicos dentro da coluna que definimos (nomecargo)
df["NOMECARGO"].unique()
```

```python
#aqui estou pegando somente valores do cargo "Analista Legislativo - AL"
analista_legislativo = df.loc[df["NOMECARGO"] == "Analista Legislativo - AL"]
analista_legislativo.head()
```

```python
#aqui eu pego quantos servidores tem em cada orgao
df.groupby("ORGAO")["Nome do servidor"].nunique()
```

```python
#convertendo coluna q esta como object para float
df["VALORLIQUIDO"] = df["VALORLIQUIDO"].apply(lambda x: str(x).replace(",","."))
df["VALORLIQUIDO"] = df["VALORLIQUIDO"].astype("float64")
```

```python
#media salarial por orgao
df.groupby("ORGAO")["VALORLIQUIDO"].mean()
```

```python
#media salarial por cargo .mean()
#soma dos salarios dos cargos
df.groupby("NOMECARGO")["VALORLIQUIDO"].sum()
```

```python
#mostrando os salarios mais altos

#axis = 0 -> exclui linha
#axis = 1 -> exclui coluna
df_sem_dados_irrelevantes = df.drop(["ANOMES", "NIVELSALARIALSECUNDARIO", "NOMECARGOSECUNDARIO", "SIMBOLOCARGOSECUNDARIO", "SIMBOLOCARGO", "NIVELSALARIAL", "VALORCORTETETO", "VALORDECIMOTERCEIRO", "VALORDEMAISDESCONTOS", "VALORFERIAS", "VALORPROVENTO", "VALORPROVENTOMES"], axis=1)
df2 = df_sem_dados_irrelevantes.sort_values(by="VALORLIQUIDO", ascending=False).head(10) #pra usar la embaixo
df_sem_dados_irrelevantes.sort_values(by="VALORLIQUIDO", ascending=False).head(10)
```

```python
#excluindo salarios menores do que 1000 reais e mostrando os menores salarios dados no csv

df_sem_salario_igual_zero = df_sem_dados_irrelevantes[(df["VALORLIQUIDO"] < 1000)].index #index pega as linha em que o valor é menor q 100

df_sem_dados_irrelevantes.drop(df_sem_salario_igual_zero, inplace=True)

df_sem_dados_irrelevantes.sort_values(by="VALORLIQUIDO", ascending=True).head(10)
```

```python
df_sem_salario_maior_que_cem_mil = df_sem_dados_irrelevantes[(df["VALORLIQUIDO"] > 100000)].index #index pega as linha em que o valor é maior que 100 mil

df_sem_dados_irrelevantes.drop(df_sem_salario_maior_que_cem_mil, inplace=True)

df_sem_dados_irrelevantes.sort_values(by="VALORLIQUIDO", ascending=False).head(10)
```

```python
#Media salarial por orgao
#o ; no final é só apagar e ver como funciona
df.groupby("Codigo do Orgao")["VALORLIQUIDO"].mean().sort_values(ascending=False).plot.barh();
```

```python
#mostrando cada código do orgao e seu orgao respectivo
orgaos = df.groupby(["Codigo do Orgao", "ORGAO"]).size().reset_index()
orgaos.columns = ['Codigo do Orgao', 'Nome do Orgao', 'Contagem']

for index, row in orgaos.iterrows():
    codigo = row['Codigo do Orgao']
    nome = row['Nome do Orgao']
    print(f"{codigo} - {nome}")
```

```python
#salvando em png
df2["ORGAO"].value_counts().plot.barh(title="Total de servidores em cada órgão\nno top 10 de salários do estado\nde Goias em abril", color="red")
plt.xlabel("Servidores")
plt.ylabel("Órgãos")
plt.legend()
plt.savefig("grafico.png")
```
